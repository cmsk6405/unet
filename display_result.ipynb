```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
result_dir = "./results/numpy"

lst_data = os.listdir(result_dir)

lst_img = [f for f in lst_data if f.startswith("img")]
lst_mask = [f for f in lst_data if f.startswith("mask")]
lst_ourput = [f for f in lst_data if f.startswith("output")]

lst_img.sort()
lst_mask.sort()
lst_ourput.sort()
```

```python
id = 0

img = np.load(os.path.join(result_dir, lst_img[id]))
mask = np.load(os.path.join(result_dir, lst_mask[id]))
output = np.load(os.path.join(result_dir, lst_ourput[id]))
```

```python
gray_image = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
plt.subplot(131)
plt.imshow(gray_image, cmap='gray')
plt.title("img")

plt.subplot(132)
plt.imshow(mask, cmap='gray')
plt.title("mask")

plt.subplot(133)
plt.imshow(output, cmap='gray')
plt.title("output")
```
